```python
x = list(range(1, 6))
y = list(range(6, 11))
print(x, y)
```

```python
plt.plot(x, y)
```

`plt.ylable`    -    title for y axis

`plt.xlable`    -    title for x axis

`plt.title`     -    title for a whole graph

```python
plt.plot(x, y)
plt.ylabel('hours')
plt.xlabel('days')
plt.title('Progress')
plt.show()
```



---



`"b"` -  blue markers with default shape

`"or"`  -  red circles

`"-g"`  -  green solid line

`"--"`  -  dashed line with default color

`"^k"`  -  black triangle_up merkers connected by dotted line

```python
plt.plot(x, y, 'b')
plt.show()
```

```python
plt.plot(x, y, 'or')
plt.show()
```

```python
plt.plot(x, y, '-g')
plt.show()
```

```python
plt.plot(x, y, '--')
plt.show()
```

```python
plt.plot(x, y, '^k')
plt.show()
```



---



`plt.axis()`  =>  (x_min, x_max, y_min, y_max)

```python
plt.plot([1, 2, 6, 7], [1, 5, 3, 9], 'or')
plt.axis((0, 10, 0, 10))
plt.show()
```



---



Column-wise breakdown:

- Column 0: values [1, 4] â†’ points (0,1) and (1,4)

- Column 1: values [2, 2] â†’ points (0,2) and (1,2)

- Column 2: values [3, 6] â†’ points (0,3) and (1,6)

- Column 3: values [4, 3] â†’ points (0,4) and (1,3)

So Matplotlib plots 4 separate lines, each connecting two points:

- Line through (0,1) â†’ (1,4)

- Line through (0,2) â†’ (1,2)

- Line through (0,3) â†’ (1,6)

- Line through (0,4) â†’ (1,3)

```python
arr = np.array([[1, 2, 3, 4], [4, 2, 6, 3]])
```

```python
plt.plot(arr)
```
