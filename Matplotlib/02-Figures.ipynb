# Visualization with MatPlotlib

Creating a figure

```python
# Importing Viz libs
import matplotlib as mpl
import matplotlib.pyplot as plt

# Import data wrangling libs

import pandas as pd
import numpy as np
```

```python
fig1 = plt.figure()
```

```python
type(fig1)
```



---



```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(2, 2, 1)
plt.show()
```

```python
fig1.patch.set_edgecolor('green')
fig1.patch.set_linewidth(2)
```

```python
fig1.figure
```



---



```python
ax2 = fig1.add_subplot(2, 2, 2)
ax3 = fig1.add_subplot(2, 2, 3)
ax4 = fig1.add_subplot(2, 2, 4)
fig1.figure
```

### Inserting graphs

```python
dataset = np.random.standard_normal(50).cumsum()
dataset
```

```python
ax1.plot(dataset)
fig1.figure
```

```python
ax2.plot(dataset, color = 'green', linestyle = '--')
fig1.figure
```

```python
ax3.plot([1, 2, 3, 4], [9, 8, 7, 6], color = 'yellow', linestyle = 'solid', alpha = 1)
fig1.figure
```

```python
data = np.random.standard_normal(100)
data
```

```python
ax4.hist(data, color = 'purple', bins = 15, alpha = 0.5)
fig1.figure
```
